```python
page = requests.get('https://www.cnbc.com/world/?region=world')

page
```

```python
soup = BeautifulSoup(page.content)
```

```python
NewsH = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    NewsH.append(i.text)
NewsH
```

```python
NewsT = []
for i in soup.find_all('span',class_="LatestNews-wrapper"):
    NewsT.append(i.text)
NewsT
```

```python
Newsl = []
a = soup.find_all("a", class_="LatestNews-headline")

a

for i in a:
    Newsl.append(i.get('href'))
Newsl
```

```python
page = requests.get('https://www.dineout.co.in/top-10-restaurants-in-delhi')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
Resname = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Resname.append(i.text)
Resname
```

```python
cuis = []
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuis.append(i.text.split('|')[1])  
    
cuis
```

```python
loca = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    loca.append(i.text)  
    
loca
```

```python
Rat = []
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Rat.append(i.text)  
    
Rat
```

```python
urla = []
img = soup.find_all("img", class_="no-img")

img

for i in img:
    urla.append(i.get('data-src'))
urla
```

```python
import pandas as pd
df = pd.DataFrame({'Restaurant name':Resname,'Cuisine':cuis,'Location':loca,'Ratings':Rat,'Image URL':urla})
df
```
