```python
import selenium 
import pandas as pd 
from selenium import webdriver 
import warnings 
warnings.filterwarnings("ignore")
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By 
import requests
import time
import re
```

```python
driver.get("https://www.amazon.in/")
```

```python
driver.maximize_window()
```

```python
#user input is ‘guitar’
guiter = driver.find_element(By.ID,"twotabsearchtextbox")
guiter.send_keys("guiter")
```

```python
guitersearch = driver.find_element(By.XPATH,'//div[@class="nav-search-submit nav-sprite"]')
guitersearch.click()
```

```python
#scrap urls
product_urls=[]
start=0
end=3
for page in range (start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-outline"]')
    for i in url[0:10]:
        product_urls.append(i.get_attribute("href"))
    driver.execute_script('window.scrollTo(0,4500);')
    nxt_button=driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
    nxt_button.click()
    time.sleep(3)
```

```python
len(product_urls)
```

```python
Brand_name=[]
Product_name=[]
prod_price=[]
delivery=[]
availability=[]

for url in product_urls:
    try:
        driver.get(url)
        time.sleep(4)
        try:
            brand=driver.find_elements(By.XPATH,'//*[@id="productOverview_feature_div"]/div/table//tr[1]/td[2]')
            for i in brand:
                bname=i.text
                Brand_name.append(bname)
        except NoSuchElementException:
            Brand_name.append("-")

        try:
            product=driver.find_elements(By.XPATH,'//span[@class="a-size-large product-title-word-break"]')
            for i in product:
                bproduct=i.text
                Product_name.append(bproduct)
        except NoSuchElementException:
            Product_name.append("-")

        try:
            proprice = driver.find_elements(By.XPATH,'//span[@class="a-price aok-align-center reinventPricePriceToPayMargin priceToPay"]')
            for i in proprice:
                bpprice=i.text
                prod_price.append(bpprice)
        except NoSuchElementException:
            prod_price.append("-")
            
        try:
            delivery_tag = driver.find_elements(By.XPATH,'//*[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]/span/span')
            for i in delivery_tag:
                bdelivery=i.text
                delivery.append(bdelivery)
        except NoSuchElementException:
            delivery.append("-")
            
        try:
            availability_tag = driver.find_elements(By.XPATH,'//*[@id="availability"]/span')
            for i in availability_tag:
                bavailability=i.text
                availability.append(bavailability)
        except NoSuchElementException:
            availability.append("-")
    except:
        pass
```

```python
len(Brand_name), len(Product_name), len(prod_price), len(delivery), len(availability), len(product_urls)
```

```python
df = pd.DataFrame({'Product Name':Product_name,'Product Price':prod_price, 'Availability':availability, 'Product URL':product_urls})
df
```

```python
letsgo = driver.find_element(By.XPATH,'//*[@id="search-button"]')
letsgo.click()
```

```python
hostel_name=[]
distance_l=[]
ratings_l=[]
description_h=[]
privates_price=[]
dorms_price=[]

hostel_tag = driver.find_elements(By.XPATH,'//*[@class="title title-6"]')
for i in hostel_tag:
    hostel = i.text
    hostel_name.append(hostel)

distance_tag = driver.find_elements(By.XPATH,'//*[@id="__layout"]/div/div[2]/div[4]/div/div//a/span[1]/span')
for i in distance_tag:
    distance = i.text
    distance_l.append(distance.replace('-',' '))

ratings_tag = driver.find_elements(By.XPATH,'//div[@class="rating rating-summary-container big"]')
for i in ratings_tag:
    rating = i.text
    ratings_l.append(rating.replace('\n',' - '))

privates_price_tag = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[0::2]
for i in privates_price_tag:
    privatesprice = i.text
    privates_price.append(privatesprice.replace('\n',': '))

dorms_price_tag = driver.find_elements(By.XPATH,'//div[@class="price-col"]')[1::2]
for i in dorms_price_tag:
    dormsprice = i.text
    dorms_price.append(dormsprice.replace('\n',': '))

    
description_tag = driver.find_elements(By.XPATH,'//*[@class="rating-factors prop-card-tablet rating-factors small"]')
for i in description_tag:
    description = i.text
    description_h.append(description.replace('\n',', '))
```

```python
len(hostel_name), len(distance_l), len(ratings_l), len(dorms_price), len(privates_price), len(description_h)
```

```python
df = pd.DataFrame({'Hostel Name':hostel_name, 'Distance from city centre':distance_l, 'Rating and review':ratings_l, 'Price Privates From':privates_price, 'Price Dorms From':dorms_price, 'Hostel Description':description_h,})
df
```
