Question 1 - Scrape the details of most viewed videos on YouTube from Wikipedia

```python
driver = webdriver.Chrome(r"C:\Users\hp\chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
driver.maximize_window()
```

```python
Rank = []
Name = []
Artist = []
Upload_date = []
Views = []

Rank_tag=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
for i in Rank_tag:
    wrank=i.text
    Rank.append(wrank)

Name_tag=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
for i in Name_tag:
    wname=i.text
    Name.append(wname)

Artist_tag=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
for i in Artist_tag:
    wartist=i.text
    Artist.append(wartist)
    
Upload_date_tag=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')
for i in Upload_date_tag:
    wupload=i.text
    Upload_date.append(wupload)

Views_tag=driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
for i in Views_tag:
    wviewstag=i.text
    Views.append(wviewstag)
```

```python
len(Rank), len(Name), len(Artist), len(Upload_date), len(Views)
```

```python
df = pd.DataFrame({'Rank':Rank,'Video Name':Name, 'Artist':Artist, 'Upload date':Upload_date, 'Views (billions)':Views})
df
```

Question 4 - Scrape the details of trending repositories on Github.com

```python
driver.get("https://github.com/")
```

```python
source = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
source.click()
trending = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a')
trending.click()
```

```python
repository_urls=[]
for page in range(0,1):
    url=driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/h1/a')
    for i in url:
        repository_urls.append(i.get_attribute("href"))
```

```python
len(repository_urls)
```

```python
Repo_title = []

Repo_title_tag=driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]//h1/a')
for i in Repo_title_tag:
    Repository=i.text
    Repo_title.append(Repository)
```

```python
len(Repo_title)
```

```python
Description = []
Contributor = []
Language = []
```

```python
#Scraping data from every repository page

for url in repository_urls:
    driver.get(url)
    time.sleep(3)
 

    try:
        desc = driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Description.append(desc.text)
    except NoSuchElementException:
        Description.append('-')
    
    try:
        Contri = driver.find_element(By.XPATH,'//*[contains(text(),"Contributors")]')
        Contributor.append(Contri.text)
    except NoSuchElementException:
        Contributor.append('-')
        
    try:
        Lang = driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        Language.append(Lang.text)
    except NoSuchElementException:
        Language.append('-')
```

```python
len(Repo_title), len(Description), len(Contributor), len(Language), 
```

```python
df = pd.DataFrame({'Repository title':Repo_title,'Repository description':Description, ' Contributors count':Contributor, ' Language used':Language,})
df
```
